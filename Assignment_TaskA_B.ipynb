```python
# Task B
def inputFile(file):
    data = []
    files = open(file, "r")
    reader = csv.reader(files)
    
    keys = []
    is_key = True
    for row in reader:
        if is_key:
            for i in range(len(row)):
                keys.append(row[i].strip())
            is_key = False
            continue
        else:
            record = {}
            for i in range(len(row)):
                try:
                    record[keys[i]] = int(row[i].strip())
                except ValueError:
                    try:
                        record[keys[i]] = float(row[i].strip())
                    except ValueError:
                        record[keys[i]] = row[i].strip()
                        
            data.append(record)
    files.close()
    return data

def merge(climate, hotspot):
    
    documents = []
    climateData = list(climate)
    hotspotData = list(hotspot)
    for cli in climateData:
        document = {}
        document["date"] = cli["date"]
        newClimate = dict(cli)
        del newClimate["date"]
        document["climate"] = newClimate
        hotspot = []
        for hot in hotspotData:
            if hot["date"] == document["date"]:
                newHotspot = dict(hot)
                del newHotspot["date"]
                hotspot.append(newHotspot)
        document["hotspot"] = hotspot
        documents.append(document)
    return documents

climate = inputFile('climate_historic.csv')
hotspot = inputFile('hotspot_historic.csv')

db.hotspot_climate.drop()
db.hotspot_climate.insert_many(merge(climate, hotspot))
```
